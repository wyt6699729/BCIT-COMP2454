```python
import matplotlib.pyplot as plt

years = [2008, 2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018]
colorado = [5029196,5029316,5048281,5121771,5193721,5270482,5351218,5452107,
5540921,5615902,5695564]
connecticut = [3574097,3574147,3579125,3588023,3594395,3594915,3594783,3587509,
3578674,3573880,3572665]
delaware = [897934,897934,899595,907316,915188,923638,932596,941413,949216,
957078,967171]

plt.plot(years, colorado, '--', color='red', label='Colorado')
plt.plot(years, connecticut, 's', color='blue', label='Connecticut')
plt.plot(years, delaware, 'o', color='green', label='Delaware')

plt.ylim(ymin=0)
plt.legend(loc='lower left', ncol=3, fancybox=True, shadow=True)
plt.xlabel('Year')
plt.ylabel('Population')
plt.title('Population by State')

plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt

NUM_MEANS = 5
NUM_GROUPS = 3
bc_means = [20, 35, 30, 35, 27]
alberta_means = [25, 32, 34, 20, 25]
saskatchewan = [18, 28, 32, 24, 31]

ind = np.arange(NUM_MEANS)
width = 0.25
plt.bar(ind, bc_means, width, color='blue', label='BC')
plt.bar(ind + width, alberta_means, width, color='orange', label='AB')
plt.bar(ind + width * 2, saskatchewan, width, color='green', label='SK')

plt.ylabel('Revenue')
plt.title('Quaterly Revenue by Province')
plt.legend(loc='best')
plt.xticks(ind + width * 2 / NUM_GROUPS, ('Q1', 'Q2', 'Q3', 'Q4', 'Q5'))

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Import data into a DataFrame
path = 'datasets/bodyfat.txt'
df = pd.read_csv(path, skiprows=1, sep='\t',
                 names=('Density', 'Pct.BF', 'Age', 'Weight', 'Height',
                    'Neck', 'Chest', 'Abdomen', 'Waist', 'Hip', 'Thigh',
                    'Ankle', 'Knee', 'Bicep', 'Forearm', 'Wrist')
                 )
# Show all columns
pd.set_option('display.max_columns', None)
# Increase nuumber of columns that display on one line
pd.set_option('display.width', 1000)

# This line allows us to set the figure size supposedly in inches
# When rendered in the IDE the output often does not translate to inches
plt.subplots(nrows=2, ncols=2, figsize=(14, 7))

# Specifies total rows, columns and image where images are drawn clockwise
plt.subplot(2, 2, 1)
plt.hist(df['Pct.BF'], bins=10, color='blue')
plt.xlabel('Pct.BF')

plt.subplot(2, 2, 2)
plt.hist(df['Age'], bins=10, color='blue')
plt.xlabel('Age')

plt.subplot(2, 2, 3)
plt.hist(df['Weight'], bins=10, color='blue')
plt.xlabel('Weight')

plt.subplot(2, 2, 4)
plt.hist(df['Height'], bins=10, color='blue')

df.head()
```
